```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("../game_logs.csv")

# Plot rewards over steps
plt.figure(figsize=(10, 6))
plt.plot(data["step"], data["rewards"], label="Rewards")
plt.xlabel("Step")
plt.ylabel("Reward")
plt.title("Rewards Over Time")
plt.legend()
plt.show()
```

```python
# Plot energy levels
plt.figure(figsize=(10, 6))
plt.plot(data["step"], data["robot_1_energy"], label="Robot 1 Energy")
plt.plot(data["step"], data["robot_2_energy"], label="Robot 2 Energy")
plt.xlabel("Step")
plt.ylabel("Energy Level")
plt.title("Energy Levels Over Time")
plt.legend()
plt.show()
```

```python
# Plot packages delivered
plt.figure(figsize=(10, 6))
plt.plot(data["step"], data["packages_delivered"], label="Packages Delivered")
plt.xlabel("Step")
plt.ylabel("Number of Packages Delivered")
plt.title("Progress of Package Delivery")
plt.legend()
plt.show()
```
